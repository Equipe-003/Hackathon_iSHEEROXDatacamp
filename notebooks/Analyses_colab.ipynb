### 1- Les 50 médias qui parlent le plus du Bénin dans le monde

```python
import re
import plotly.express as px

# Fonction pour extraire le domaine de base d'une URL
def get_base_domain(url):
    try:
        # Supprimer le protocole (http/https) et le préfixe 'www.'
        domain = re.sub(r'^(https?://)?(www\.)?', '', url.lower())
        # Garder uniquement la partie avant le premier '/' si elle existe
        domain = domain.split('/')[0]
        return domain
    except:
        return None

# Appliquer la fonction pour créer une nouvelle colonne 'MediaSource'
df_filtre['MediaSource'] = df_filtre['SOURCEURL'].apply(get_base_domain)

# Compter les occurrences pour chaque source médiatique
media_counts = df_filtre['MediaSource'].value_counts().reset_index()
media_counts.columns = ['MediaSource', 'Mentions']

# Obtenir les 50 principales sources médiatiques et créer explicitement une copie pour éviter SettingWithCopyWarning
top_50_media = media_counts.head(50).copy()

# Calculer la 'part' (pourcentage)
top_50_media['Part'] = (top_50_media['Mentions'] / top_50_media['Mentions'].sum()) * 100

print("Top 50 des médias qui parlent le plus du Bénin :")
print(top_50_media[['MediaSource', 'Mentions', 'Part']])

# Créer un Treemap
fig = px.treemap(
    top_50_media,
    path=['MediaSource'],
    values='Mentions',
    title='Distribution des Mentions du Bénin par Média (Top 50)',
    color='Part',
    color_continuous_scale='Blues',
    hover_data={'MediaSource':True, 'Mentions':True, 'Part':':.2f%'} # Afficher la part avec 2 décimales
)

fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

### 2. Évolution de la Tonalité de l'Actualité sur le Bénin au cours de l'Année

```python
import pandas as pd
import plotly.express as px

# Convertir SQLDATE en format de date si ce n'est pas déjà fait
df_filtre['SQLDATE_DATE'] = pd.to_datetime(df_filtre['SQLDATE'], format='%Y%m%d')

# Extraire le mois et le nom du mois
df_filtre['Month'] = df_filtre['SQLDATE_DATE'].dt.month
df_filtre['MonthName'] = df_filtre['SQLDATE_DATE'].dt.strftime('%B')

# Calculer la tonalité moyenne par mois
tone_by_month = df_filtre.groupby(['Month', 'MonthName'])['AvgTone'].mean().reset_index()
tone_by_month = tone_by_month.sort_values('Month')

# Identifier le mois le plus positif et le plus négatif
most_positive_month = tone_by_month.loc[tone_by_month['AvgTone'].idxmax()]
most_negative_month = tone_by_month.loc[tone_by_month['AvgTone'].idxmin()]

print(f"Le mois le plus positif en termes de tonalité moyenne est : {most_positive_month['MonthName']} ({most_positive_month['AvgTone']:.2f})")
print(f"Le mois le plus négatif en termes de tonalité moyenne est : {most_negative_month['MonthName']} ({most_negative_month['AvgTone']:.2f})")

# Créer une courbe d'évolution de la tonalité moyenne
fig = px.line(
    tone_by_month,
    x='MonthName',
    y='AvgTone',
    title='Évolution de la Tonalité Moyenne de l\'Actualité sur le Bénin par Mois',
    labels={'MonthName': 'Mois', 'AvgTone': 'Tonalité Moyenne'},
    markers=True,
    line_shape='linear'
)

# Mettre en évidence le mois le plus positif et le plus négatif
fig.add_scatter(
    x=[most_positive_month['MonthName']],
    y=[most_positive_month['AvgTone']],
    mode='markers',
    name='Mois le plus positif',
    marker=dict(color='green', size=10)
)
fig.add_scatter(
    x=[most_negative_month['MonthName']],
    y=[most_negative_month['AvgTone']],
    mode='markers',
    name='Mois le plus négatif',
    marker=dict(color='red', size=10)
)

fig.update_layout(xaxis_title='Mois', yaxis_title='Tonalité Moyenne')
fig.show()
```

### 3- A quel moment le Bénin a il le plus fait parlé de lui (buzzé) dans le monde

```python
import pandas as pd
import re

# Fonction pour extraire le domaine de base d'une URL
def get_base_domain(url):
    try:
        # Supprimer le protocole (http/https) et le préfixe 'www.'
        domain = re.sub(r'^(https?://)?(www\\.)?', '', url.lower())
        # Garder uniquement la partie avant le premier '/' si elle existe
        domain = domain.split('/')[0]
        return domain
    except:
        return None

# S'assurer que la colonne 'SQLDATE_DATE' est au format datetime et extraire le mois et le nom du mois
# (Ces étapes ont été faites précédemment, mais on les répète ici pour s'assurer de la cohérence)
if 'SQLDATE_DATE' not in df_filtre.columns:
    df_filtre['SQLDATE_DATE'] = pd.to_datetime(df_filtre['SQLDATE'], format='%Y%m%d', errors='coerce')
    # Supprimer les lignes où la conversion de date a échoué (NaT)
    df_filtre.dropna(subset=['SQLDATE_DATE'], inplace=True)

df_filtre['Month'] = df_filtre['SQLDATE_DATE'].dt.month
df_filtre['MonthName'] = df_filtre['SQLDATE_DATE'].dt.strftime('%B')

# Créer la colonne 'MediaSource' si elle n'existe pas
if 'MediaSource' not in df_filtre.columns:
    df_filtre['MediaSource'] = df_filtre['SOURCEURL'].apply(get_base_domain)

# Agrégation des données par mois pour calculer les composants du score de buzz
monthly_buzz_components = df_filtre.groupby(['Month', 'MonthName']).agg(
    NumMentions=('SQLDATE', 'size'),  # Nombre total d'événements (mentions)
    NumArticles=('SOURCEURL', 'nunique'),  # Nombre d'articles uniques
    NumSources=('MediaSource', 'nunique')  # Nombre de sources uniques
).reset_index()

# Calculer le score de buzz en utilisant la formule fournie
monthly_buzz_components['BuzzScore'] = (
    0.5 * monthly_buzz_components['NumMentions'] +
    0.3 * monthly_buzz_components['NumArticles'] +
    0.2 * monthly_buzz_components['NumSources']
)

# Trier par mois (optionnel, pour l'affichage)
monthly_buzz_components = monthly_buzz_components.sort_values('Month')

# Fusionner df_filtre avec monthly_buzz_components pour ajouter le 'BuzzScore' à chaque ligne d'événement
# Utiliser une copie du df_filtre pour éviter les SettingWithCopyWarning potentielles
df_filtre = df_filtre.merge(
    monthly_buzz_components[['Month', 'MonthName', 'BuzzScore']],
    on=['Month', 'MonthName'],
    how='left'
)

```

```python
import plotly.express as px

# Identifier le mois où le Bénin a le plus buzzé
most_buzzed_month_data = monthly_buzz_components.loc[monthly_buzz_components['BuzzScore'].idxmax()]

print(f"Le Bénin a le plus buzzé en : {most_buzzed_month_data['MonthName']} avec un score de buzz de {most_buzzed_month_data['BuzzScore']:.2f}")
print("\nScore de Buzz par mois :")
print(monthly_buzz_components[['MonthName', 'BuzzScore']].to_string())

# Créer un graphique pour visualiser le score de buzz par mois
fig_buzz_score = px.bar(
    monthly_buzz_components,
    x='MonthName',
    y='BuzzScore',
    title='Score de Buzz de l\'Actualité sur le Bénin par Mois',
    labels={'MonthName': 'Mois', 'BuzzScore': 'Score de Buzz'},
    color='BuzzScore',
    color_continuous_scale='Plasma'
)

# Mettre en évidence le mois le plus buzzé
fig_buzz_score.add_scatter(
    x=[most_buzzed_month_data['MonthName']],
    y=[most_buzzed_month_data['BuzzScore']],
    mode='markers',
    name='Mois le plus buzzé',
    marker=dict(color='red', size=12, symbol='star')
)

fig_buzz_score.update_layout(xaxis_title='Mois', yaxis_title='Score de Buzz')
fig_buzz_score.show()
```

L'actualité béninoise a connu un grand buzz en décembre. Nous avons analysé le « Score de Buzz » pour identifier les moments et les sujets qui ont généré le plus d'attention médiatique concernant le Bénin.

#### Période de plus grand buzz

### Visualisation des Tendances Mensuelles : Buzz et Tonalité

```python
import pandas as pd
import re
import plotly.graph_objects as go


# Ensure 'SQLDATE_DATE' is in datetime format and 'Month'/'MonthName' exist
# Use pd.api.types.is_datetime64_any_dtype to robustly check date type, or convert if not present/incorrect type
if 'SQLDATE' not in df_filtre.columns or not pd.api.types.is_datetime64_any_dtype(df_filtre['SQLDATE_DATE']):
    df_filtre['SQLDATE'] = pd.to_datetime(df_filtre['SQLDATE'], format='%Y%m%d', errors='coerce')
    df_filtre.dropna(subset=['SQLDATE'], inplace=True)

# Ensure Month and MonthName columns are present in df_filtre
if 'Month' not in df_filtre.columns or 'MonthName' not in df_filtre.columns:
    df_filtre['Month'] = df_filtre['SQLDATE'].dt.month
    df_filtre['MonthName'] = df_filtre['SQLDATE'].dt.strftime('%B')

# Function to extract the base domain from a URL
def get_base_domain(url):
    try:
        # Supprimer le protocole (http/https) et le préfixe 'www.'
        domain = re.sub(r'^(https?://)?(www\\.)?', '', url.lower())
        # Garder uniquement la partie avant le premier '/' si elle existe
        domain = domain.split('/')[0]
        return domain
    except:
        return None

# Create the 'MediaSource' column if it doesn't exist
if 'MediaSource' not in df_filtre.columns:
    df_filtre['MediaSource'] = df_filtre['SOURCEURL'].apply(get_base_domain)

# Agrégation des données par mois pour calculer les composants du score de buzz
monthly_buzz_components = df_filtre.groupby(['Month', 'MonthName']).agg(
    NumMentions=('SQLDATE', 'size'),  # Nombre total d'événements (mentions)
    NumArticles=('SOURCEURL', 'nunique'),    # Nombre d'articles uniques
    NumSources=('MediaSource', 'nunique')   # Nombre de sources uniques
).reset_index()

monthly_buzz_components['BuzzScore'] = (
    0.5 * monthly_buzz_components['NumMentions'] +
    0.3 * monthly_buzz_components['NumArticles'] +
    0.2 * monthly_buzz_components['NumSources']
)

# Ensure months are in chronological order for plotting
# Need to use the numeric Month for sorting then set categorical for display
monthly_buzz_components['MonthName'] = pd.Categorical(monthly_buzz_components['MonthName'],
                                                  categories=['January', 'February', 'March', 'April', 'May', 'June',
                                                              'July', 'August', 'September', 'October', 'November', 'December'],
                                                  ordered=True)
monthly_buzz_components = monthly_buzz_components.sort_values('MonthName') # Sort by the categorical MonthName

# Identify the most buzzed month
most_buzzed_month_data = monthly_buzz_components.loc[monthly_buzz_components['BuzzScore'].idxmax()]


# Graphique pour le Score de Buzz
fig_buzz_area = go.Figure()

fig_buzz_area.add_trace(go.Scatter(
    x=monthly_buzz_components['MonthName'],
    y=monthly_buzz_components['BuzzScore'],
    mode='lines+markers',
    fill='tozeroy', # Fill the area below the line
    name='Score de Buzz',
    line=dict(color='blue', width=2),
    marker=dict(color='blue', size=8)
))

# Mettre en évidence le mois le plus buzzé
fig_buzz_area.add_trace(go.Scatter(
    x=[most_buzzed_month_data['MonthName']],
    y=[most_buzzed_month_data['BuzzScore']],
    mode='markers',
    name='Mois le plus buzzé',
    marker=dict(color='red', size=12, symbol='star', line=dict(color='black', width=1))
))

fig_buzz_area.update_layout(
    title='Évolution du Score de Buzz par Mois (Area Chart)',
    xaxis_title='Mois',
    yaxis_title='Score de Buzz',
    hovermode='x unified'
)

fig_buzz_area.show()
```

### 5 -  Analyse des 4 grandes catégories d'événements en fonction du temps


```python
import plotly.express as px

# S'assurer que les colonnes de date existent et sont au bon format, et que df_filtre a les colonnes nécessaires
# Ces étapes sont reprises pour robustesse, au cas où le kernel aurait été redémarré ou que des colonnes auraient été modifiées.
if 'SQLDATE_DATE' not in df_filtre.columns or not pd.api.types.is_datetime64_any_dtype(df_filtre['SQLDATE_DATE']):
    df_filtre['SQLDATE_DATE'] = pd.to_datetime(df_filtre['SQLDATE'], format='%Y%m%d', errors='coerce')
    df_filtre.dropna(subset=['SQLDATE_DATE'], inplace=True)

if 'Month' not in df_filtre.columns or 'MonthName' not in df_filtre.columns:
    df_filtre['Month'] = df_filtre['SQLDATE_DATE'].dt.month
    df_filtre['MonthName'] = df_filtre['SQLDATE_DATE'].dt.strftime('%B')

# Définir un mapping pour les valeurs numériques de 'QuadClass' vers des descriptions en français
quadclass_map = {
    1: 'Coopération Verbale',
    2: 'Coopération Matérielle',
    3: 'Conflit Verbal',
    4: 'Conflit Matériel'
}

# Créer la colonne 'QuadClassDescription' si elle n'existe pas ou la mettre à jour
if 'QuadClassDescription' not in df_filtre.columns:
    df_filtre['QuadClassDescription'] = df_filtre['QuadClass'].map(quadclass_map)

# Grouper le DataFrame par 'Month', 'MonthName' et 'QuadClassDescription', puis compter les occurrences
quadclass_monthly_counts = df_filtre.groupby(['Month', 'MonthName', 'QuadClassDescription']).size().reset_index(name='Count')

# Calculer le total des événements par mois
monthly_total_counts = quadclass_monthly_counts.groupby(['Month', 'MonthName'])['Count'].transform('sum')

# Calculer les proportions mensuelles
quadclass_monthly_proportions = quadclass_monthly_counts.copy() # Faire une copie pour éviter les problèmes de SettingWithCopyWarning
quadclass_monthly_proportions['Proportion'] = (100 * quadclass_monthly_proportions['Count'] / monthly_total_counts)

# Définir l'ordre des mois pour un affichage chronologique et traduire en français
month_order_english = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
month_order_french = ['Janvier', 'Février', 'Mars', 'Avril', 'Mai', 'Juin', 'Juillet', 'Août', 'Septembre', 'Octobre', 'Novembre', 'Décembre']

# Créer un mapping pour traduire les noms de mois
month_translation_map = dict(zip(month_order_english, month_order_french))
quadclass_monthly_proportions['MonthName_FR'] = quadclass_monthly_proportions['MonthName'].map(month_translation_map)

# Convertir 'MonthName_FR' en catégorie pour un tri correct
quadclass_monthly_proportions['MonthName_FR'] = pd.Categorical(quadclass_monthly_proportions['MonthName_FR'], categories=month_order_french, ordered=True)
quadclass_monthly_proportions = quadclass_monthly_proportions.sort_values(['MonthName_FR', 'QuadClassDescription'])

# Créer le graphique en aires empilées à 100%
fig_quadclass_stacked_area = px.area(
    quadclass_monthly_proportions,
    x='MonthName_FR',
    y='Proportion',
    color='QuadClassDescription',
    title='Distribution Mensuelle des Types d\'Événements (QuadClass) (100% Empilé)',
    labels={'MonthName_FR': 'Mois', 'Proportion': 'Proportion (%)', 'QuadClassDescription': 'Type d\'Événement'},
    groupnorm='percent', # Cela assure que les aires sont empilées à 100%
    category_orders={'MonthName_FR': month_order_french},
    color_discrete_map={
        'Coopération Verbale': 'lightblue',
        'Coopération Matérielle': 'blue',
        'Conflit Verbal': 'orange',
        'Conflit Matériel': 'red'
    }
)

fig_quadclass_stacked_area.update_layout(xaxis_title='Mois', yaxis_title='Proportion (%)')
fig_quadclass_stacked_area.show()
```
